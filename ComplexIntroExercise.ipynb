## Exercise 2: Visualizing Complex Operations
1. Write a Python function to plot a complex number on the complex plane. Use it to plot `(3 + 4j)` and its conjugate.
2. Extend the function to show addition and multiplication of two complex numbers graphically.


```python
import matplotlib.pyplot as plt
import numpy as np

def plot_complex_number(z, label):
    plt.plot(z.real, z.imag, 'o', label=label)

n = 3 + 4j
n_conjugate = n.conjugate()

n1 = 3 + 2j
n2 = 4 + 1j

addi = n1 + n2
mult = n1 * n2

# Graficar 
plot_complex_number(addi, f'suma: {addi}' )
plot_complex_number(mult, f'multiplicacion: {mult}')
plot_complex_number(n, '3 + 4j')
plot_complex_number(n_conjugate, 'Conjugado: 3 - 4j')

# Configuraciones del gráfico
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginaria')
plt.legend()
plt.title('Número Complejo y su Conjugado')
plt.xlim(-2 , 16)
plt.ylim(-6 , 16)
plt.show()
```


## Exercise 3: Exploring the Mandelbrot Set
1. Modify the provided Mandelbrot set code to change its zoom level and center point. Observe how the fractal pattern changes.
2. Experiment with different values of `max_iter` and observe the effect on the fractal's detail and computation time.





```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Parameters
pixels = 800
max_iter = 100  # Increased max_iter for more detail
zoom_level = 0.5  # Adjust zoom level (smaller values zoom in)
center_x = -0.5  # Center point x-coordinate
center_y = 0.0   # Center point y-coordinate

# Create a grid of complex numbers with adjusted zoom and center
#x.min() será center_x - zoom_level = -0.5 - 0.5 = -1.0.
#x.max() será center_x + zoom_level = -0.5 + 0.5 = 0.0.
#y.min() será center_y - zoom_level = 0.0 - 0.5 = -0.5.
#y.max() será center_y + zoom_level = 0.0 + 0.5 = 0.5.
x = np.linspace(center_x - zoom_level, center_x + zoom_level, pixels)
y = np.linspace(center_y - zoom_level, center_y + zoom_level, pixels)
xx, yy = np.meshgrid(x, y)
c = xx + yy*1j

# Compute Mandelbrot set
mandelbrot_set = np.array([mandelbrot(ci, max_iter) for ci in c.ravel()]).reshape(c.shape)

# Plot
plt.imshow(mandelbrot_set, extent=[x.min(), x.max(), y.min(), y.max()])
plt.colorbar()
plt.title(f"Mandelbrot Set (Zoom: {zoom_level}, Center: ({center_x}, {center_y}))")
plt.show()
```


## Exercise 4: Creating a Julia Set
1. Implement a Julia set generator. Use a constant like `-0.4 + 0.6j` for the iteration `z = z*z + constant`.
2. Explore how changing the constant changes the pattern of the Julia set.




```python
import numpy as np
import matplotlib.pyplot as plt

def julia_set(width, height, c, x_range=(-2, 2), y_range=(-2, 2), max_iter=100):
    # Matriz para almacenar los resultados
    julia = np.zeros((width, height))

    #  ejes x y en el plano complejo
    x = np.linspace(x_range[0], x_range[1], width)
    y = np.linspace(y_range[0], y_range[1], height)

    # Iterar sobre cada punto en el plano complejo
    for i in range(width):
        for j in range(height):
            z = x[i] + y[j]*1j  # Convertir a número complejo
            iteration = 0

            # función z = z^2 + c
            while abs(z) <= 2 and iteration < max_iter:
                z = z*z + c
                iteration += 1

            # Almacena el número de iteraciones en la matriz
            julia[i, j] = iteration

    return julia

# Parámetros conjunto de Julia
width, height = 800, 800
c = -0.4 + 0.6j

julia = julia_set(width, height, c)

plt.figure(figsize=(10, 10))
plt.imshow(julia.T, cmap='hot', extent=(-2, 2, -2, 2))
plt.colorbar(label='Iteraciones')
plt.title(f'Conjunto de Julia para c = {c}')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```
